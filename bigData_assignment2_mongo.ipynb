```python
#Query number 1 and 3
#Insert the data into the Database and remove the three columns

taxi_trips = []
for x in range(0,2000):
    taxi_trips = taxi_trips + [{'vendor_id': int(trip_df.loc[x].vendor_id),
                                'pickup_datetime': datetime.strptime(trip_df.loc[x].pickup_datetime, '%Y-%m-%d %H:%M:%S'),
                                'dropoff_datetime': datetime.strptime(trip_df.loc[x].dropoff_datetime, '%Y-%m-%d %H:%M:%S'),
                                'passenger_count': int(trip_df.loc[x].passenger_count), 
                                'trip_distance': float(trip_df.loc[x].trip_distance), 
                                #'rate code': '%s' % (trip_df.loc[x].rate_code), 
                                #'store and fwd flag': '%s' % (trip_df.loc[x].store_and_fwd_flag), 
                                'payment_type': int(trip_df.loc[x].payment_type), 
                                'fare_amount': float(trip_df.loc[x].fare_amount), 
                                'extra': float(trip_df.loc[x].extra), 
                                'mta_tax': float(trip_df.loc[x].mta_tax), 
                                'tip_amount': float(trip_df.loc[x].tip_amount), 
                                'tolls_amount': float(trip_df.loc[x].tolls_amount), 
                                'imp_surcharge': float(trip_df.loc[x].imp_surcharge), 
                                #'total amount': '%d' % (trip_df.loc[x].total_amount), 
                                'pickup_location_id': int(trip_df.loc[x].pickup_location_id), 
                                'dropoff_location_id': int(trip_df.loc[x].dropoff_location_id),
                               }]
collection.insert_many(taxi_trips)
```

```python
#Query number 4
#Calculate the duration

duration=collection.aggregate([
    {
        "$project":{
            "item":1, "duration":{
                "$divide": [{
                    "$subtract":["$dropoff_datetime", "$pickup_datetime"]
                }, 1000*60 ] 
            }
        }
    },
])
dur=[]
for i in duration:
    print("duration is",i['duration'])
    dur= dur+[{'duration': i['duration']}]
d = pd.DataFrame(dur)
```

```python
#Query number 5
#Calculate the trip cost

trip_cost = collection.aggregate(
    [
        { 
            "$project" : {
                "total_trip_cost" : { 
                    "$add": ["$extra", "$mta_tax", "$tip_amount", "$tolls_amount", "$imp_surcharge"]
                }
            }
        }
    ]
)
#collection.insert_many(trip_cost)
tr_cost = []
for i in trip_cost:
    print("total trip cost is: ",i['total_trip_cost'])
    tr_cost= tr_cost+[{'trip_cost': i['total_trip_cost']}]
trip_c = pd.DataFrame(tr_cost)
```

```python
#read the data again with parse_dates

df = pd.read_csv('D:/computer science GIU/semester 6/BigData/GIU_2718_56_7524_2022-05-17T10_13_53/datasets/taxi_trip_data.csv', encoding= 'unicode_escape', parse_dates=["pickup_datetime", "dropoff_datetime"]) # edit path accordingly to read from where the listings.csv file is
i,j = df.shape
```

```python
#Query number 6
#the most common payment type

morning = []
afternoon = []
evening = []
for i in range (0, 2000):
    if(df.loc[i].pickup_datetime.hour >= 5 and df.loc[i].pickup_datetime.hour < 12):
        morning.append(df.iloc[i]['payment_type'])
    if(df.loc[i].pickup_datetime.hour >= 12 and df.loc[i].pickup_datetime.hour < 17):
        afternoon.append(df.iloc[i]['payment_type'])
    if (df.loc[i].pickup_datetime.hour>= 17 and df.loc[i].pickup_datetime.hour < 21):
            evening.append(df.iloc[i]['payment_type'])
```

```python
# 1=Credit_card 2=Cash 3=No_charge 4=Dispute 5=Unknown 6=Voidedtrip

a = np.array(morning)
unique, counts = np.unique(a, return_counts=True)
dict(zip(unique, counts))
```

```python
#Query number 7
#Calculate the tip average
tip = collection.aggregate(
    [
        {
            '$project': {
                "tip_avg": {
                    '$avg':["$tip_amount"]
                }
            }
        },
    ]
)
for l in tip:
    print("tip average is: ",l['tip_avg'])
```

```python
#Query number 8
#Calculate the best 5 locations
top_5 = collection.aggregate(
    [
        {
            "$group": {
                "_id": {"pickup_location_id" : "$pickup_location_id"},
                 "count": {"$sum": 1} }
        },
        {
            "$sort" : { "count" : -1 } 
        },
        {
            "$limit" : 5
        },
    ]
)
print("the best 5 locations are:")
for l in top_5:
    print(l['_id'],"count:",l['count'])
```
